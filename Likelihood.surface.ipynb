# Analysis of likelihood mode
A run on the mode {1.52381435573e-06, 0.00105642140151, 0.000102755080743, 0.233992243545, 1.4362503806e-06, 4.18286907053e-07, 0.000153748654986, 0.00158103916315, 0.267335608114, 0.000792355496284, 1.0, 0.0682803227911, 0.00714847521961, 0.0, 1.04092298604} was performed. Here we see the trajectories. The mode was initially found with the [script](../workflows/workflow-ms_generated.py) on the [version at the time](https://github.com/mailund/IMCoalHMM/tree/6107264fbec31d455cbe696cb58a00516ab5cb28). The script makes two different simulations; one with scaling and one without. This mode is from the one without. 

```r
numbers=read.csv("../Data//mode_pulse_log_likelihood.txt", header=F)
names_pulse=c('1c1', '1c2','1c3','1c4','2c1','2c2','2c3','2c4','12m1', '12m2','12m3','21m1','21m2','21m3','rho')
rownames(numbers) <- names_pulse
```

```r
library(xtable)
xtable(numbers)
```

```r
mode=c(1.52381435573e-06, 0.00105642140151, 0.000102755080743, 0.233992243545, 1.4362503806e-06, 4.18286907053e-07, 0.000153748654986, 0.00158103916315, 0.267335608114, 0.000792355496284, 1.0, 0.0682803227911, 0.00714847521961, 0.0, 1.04092298604)
for(i in 1:nrow(numbers)){
    plot(mode[i]*((0:39)/19+0.2),numbers[i,], type='l', xlab=names_pulse[i], ylab="likelihood")
    abline(v=mode[i], col='red')
}
```
